**Лабораторная работа №3. Метод Монте-Карло.**

```python
import math
import random
```

## а)

```python
def g_1(x):
    return 1 / (x ** 2 - 2 * x * math.cos(math.pi / 7) + 1)
```

```python
n = 1_000_000
a = 0

l = -1
r = 1

for _ in range(n):
    eta = random.uniform(l, r)
    a += g_1(eta)

integral_approx = a * (r - l) / n
print('Приближенное значение интеграла:', integral_approx)

integral_value = math.pi / math.sqrt(2 - 2 * math.sin(3 * math.pi / 14))
print('Точное значение интеграла:', integral_value)
```

## б)

```python
def g_2(t):
    return 1 / (3 * t * (1 / t - 8) ** (2 / 3))
```

```python
n = 1_000_000
a = 0

l = 0
r = 1 / 8

for _ in range(n):
    eta = random.uniform(l, r)
    a += g_2(eta)

integral_approx = a * (r - l) / n
print('Приближенное значение интеграла:', integral_approx)

integral_value = math.pi / (6 * math.sqrt(3))
print('Точное значение интеграла:', integral_value)
```

```python
def g_3(x, y):
    return math.sqrt(1 + math.sin(x + 2 * y) ** 2)
```

```python
n = 1_000_000
N = 48
m = 0
s2 = 1

a_outer = 0

for _ in range(n):
    a = 0
    for _ in range(N):
        a += random.uniform(0, 1)
    eta_1 = (a - (1 / 2) * N) / (math.sqrt(1 / 12) * math.sqrt(N))
    
    a = 0
    for _ in range(N):
        a += random.uniform(0, 1)
    eta_2 = (a - (1 / 2) * N) / (math.sqrt(1 / 12) * math.sqrt(N))

    a_outer += g_3(eta_1, eta_2)

integral_approx = a_outer / n * math.pi * 2
print('Приближенное значение интеграла:', integral_approx)

integral_value = 7.64034
print('Точное значение интеграла:', integral_value)
```
